## Top 10 Players of the matches

```python
top_players = match_df.player_of_match.value_counts()[:10]
fig, ax = plt.subplots(figsize=(16,10))
ax.set_xlim([0,25])
ax.set_xlabel("Count")
ax.set_title("Top 10 player of the matches")
sns.barplot(x = top_players, y = top_players.index , orient = "h", palette="Reds");
plt.show()
```

As we see that __CH Gayle__(Cris Gayle) is the most successful player of all the matches and __Virat Kohli__ is at number 10 

### Number of matches played by each team:

```python
df1 = pd.melt(match_df, id_vars=['id','season'], value_vars=['team1', 'team2'])
sns.set_theme(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(y='value', data=df1)
plt.yticks(rotation='0')
plt.show()
```

* here we find that __Mumbai indians__ have played the __most__ matches and __Kochi tuskers Kerala__ has played the __least__ number of matches

### Number of winning by teams :-

```python
plt.figure(figsize=(18,9))
sns.set_theme(style="darkgrid")
sns.countplot(y='winner', data=match_df)
plt.yticks(rotation=0)
plt.show()
```

* as we can see __Delhi Daredevil__ has won the most number of matches and __Rising Pune Supergiants__ has won the least number of matches

### What is the winning percentage of the team when they choose to bat and bowl :-

```python
num_of_wins = (match_df.win_by_wickets>0).sum()
num_of_loss = (match_df.win_by_wickets==0).sum()
labels = ["Wins", "Loss"]
total = float(num_of_wins + num_of_loss)
sizes = [(num_of_wins/total)*100, (num_of_loss/total)*100]
colors = ['Green', 'Red']
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Win percentage when teams bat second")
plt.show()
```

__Top ipl run scorer__

```python
temp_df = delvry_df.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['batsman'])
ind = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(temp_df['batsman_runs']), width=0.8, color='Green')
ax.set_xticks(ind+((0.8)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top run scorers in IPL")
ax.set_xlabel('Batsmane Name')
plt.show()
```

__batsman with most number of boundries__

```python
temp_df = delvry_df.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['batsman'])
ind = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(temp_df['batsman_runs']), width=0.8, color='lightskyblue')
ax.set_xticks(ind+((0.8)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Batsman with most number of boundaries",fontsize = 10)
plt.show()
```

__Batsman with most number of sixes__

```python
temp_df = delvry_df.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['batsman'])
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(np.arange(len(labels)), np.array(temp_df['batsman_runs']), width=0.8, color='Red')
ax.set_xticks(np.arange(len(labels))+((0.8)/2.))
ax.set_xticklabels(labels, rotation=45)
ax.set_ylabel("Count")
ax.set_title("Batsman with most number of sixes.!")
ax.set_xlabel('Batsmane Name')
plt.show()
```

## Bowler Analysis:-

__Bowler with most number of wickets__

```python
temp_df = delvry_df.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['bowler'])
ls = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ls, np.array(temp_df['ball']), width=0.8, color='cyan')
ax.set_xticks(ls+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top 10 Bowlers with Number of balls bowled in IPL")
ax.set_xlabel('Bowler Names')
plt.show()
```

__Bowlers with most number of dot balls__

```python
temp_df = delvry_df.groupby('bowler')['total_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='total_runs', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['bowler'])
ls = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ls, np.array(temp_df['total_runs']), width=0.8, color='Lightpink')
ax.set_xticks(ls+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top 10 Bowlers with Number of dot balls bowled in IPL")
ax.set_xlabel('Bowler Names')
plt.show()
```
